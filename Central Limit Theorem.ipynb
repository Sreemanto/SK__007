## Central limit theorem says that with a large enough collection of samples from the same population, the sample means will be normally distributed. Note that this doesn't make any assumptions about the underlying distribution of the data; with a reasonably large sample of roughly 30 or more, this theorem will always ring true no matter what the population looks like.

```python
# Create a sample of 10 die rolls
small = randint(1, 7, 10)
print(small)
# Calculate and print the mean of the sample
small_mean = small.mean()
print(small_mean)

# Create a sample of 1000 die rolls
large = randint(1, 7, 1000)

# Calculate and print the mean of the large sample
large_mean = large.mean()
print(large_mean)

# Create a sample of 100000 die rolls
verylarge = randint(1, 7, 100000)

verylarge = verylarge.mean()
print(verylarge)

verylarge2 = randint(1, 7, 1000000)

verylarge2 = verylarge2.mean()
print(verylarge2)

print("Notice how the mean of the large sample has gotten closer to the true expected mean value of 3.5 for a rolled die.")
```

### Create and show a histogram of the means using the hist() function; examine the shape of the distribution.

```python
import matplotlib.pyplot as plt
```

```python
means = [randint(1, 7, 30).mean() for i in range(1000)]

# Create and show a histogram of the means
plt.hist(means)
plt.show()
```

```python
# Adapt code for 100 samples of size 30
means = [randint(1, 7, 30).mean() for i in range(200)]

# Create and show a histogram of the means
plt.hist(means)
plt.show()
```

#### Note how whether we took 100 or 1000 sample means, the distribution was still approximately normal. This will always be the case when we have a large enough sample (typically above 30). That's the central limit theorem at work. Remember why it's so important. It serves as the basis for all statistical experiments that you'll do!
