# Import all the required Packages

```python
import pandas as pd
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from IPython.display import Image
```

# Load the Dataset 

```python
df = pd.read_csv('C:/Users/sreemanta.kesh/Desktop/Clarity/Poll_Survey.csv')
```

```python
df.head(5)
```

```python
df.shape
```

# Drop the columns that are not required

```python
df = df.drop(['Timestamp','Unnamed: 0'],axis=1)
```

# Replace the columns with numbers

```python
df = df.replace(['Female','Male'],[0,1])
df = df.replace(['0-18','18-30','30-50','50-60','60+'],[1,2,3,4,5])
df = df.replace(['Marmara','Ege','Akdeniz','İç Anadolu','Karadeniz','Doğu Anadolu','Güneydoğu'],[1,2,3,4,5,6,7])
df = df.replace(['University','High School','Associate Degree','MA','Primary High School','Junior High School'],[1,2,3,4,5,6])
df = df.replace(['Yes','No'],[1,0])
df = df.replace(['IYI PARTI','CHP','AKP','HDP','MHP','OTHERS'],[1,2,3,4,5,6])
```

```python
df.head(5)
```

# Specify the Feature and Target 

```python
X = df.iloc[ : , 0:14]
Y = df.iloc[ : ,14:]
```

# Define the Classifier (specify the max_depth and max_leaf_nodes)

```python
Tree = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=4,
            max_features=None, max_leaf_nodes=10,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, presort=False, random_state=None,
            splitter='best')
```

# Train the Dataset

```python
Tree.fit(X,Y)
```

# Use export_graphviz to generate the Tree in .dot file

```python
Feature = ['Gender','Age_Group','City','Education','Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10']
Target = ['IYI PARTI','CHP','AKP','HDP','MHP','OTHERS']
```

```python
export_graphviz(Tree, out_file='Diagram.dot', feature_names = Feature,
                class_names = Target,
                rounded = True, proportion = False, precision = 2, filled = True)
```
