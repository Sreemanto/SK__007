### Rename the Dataframe 

```python
df = df.rename({'Cinsiyet':'Gender','Yas':'Age_Group','Bolge':'City','Egitim':'Education','soru1':'Q1','soru2':'Q2','soru3':'Q3','soru4':'Q4','soru5':'Q5','soru6':'Q6','soru7':'Q7','soru8':'Q8','soru9':'Q9','soru10':'Q10','parti':'Party'},axis=1)
```

### Replace all the repeating Turkish words by thier corresponding English meaning at once 

```python
df = df.replace(['Erkek','Kadın','Hayır','Evet','DIĞER'],['Male','Female','No','Yes','OTHERS'])
```

```python
df.head(5)
```

### Do the same for Education columns 

```python
df = df.replace(['Lisans','Lise','Lisans Üstü','Ön Lisans','İlkokul','Ortaokul'],['University','High School','MA','Associate Degree','Primary High School','Junior High School'])
```
