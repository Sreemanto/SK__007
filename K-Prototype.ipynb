```python
import numpy as np
import pandas as pd

from kmodes.kprototypes import KPrototypes
```

```python
df = pd.read_csv('Marketing.csv')
```

```python
df.columns
```

```python
df
```

```python
df=df.drop(['Customer','Vehicle_Class','avg_vehicle_age','months_last_claim','Total_Claim_Amount'],axis=1)
```

```python
df.head(3)
```

```python
df_array=df.values
```

```python
df_array[:, 1] = df_array[:, 1].astype(float)
df_array[:, 3] = df_array[:, 3].astype(float)
df_array[:, 5] = df_array[:, 5].astype(float)
df_array[:, 6] = df_array[:, 6].astype(float)
```

```python
kproto = KPrototypes(n_clusters=3, verbose=2,max_iter=20)
```

```python
clusters = kproto.fit_predict(df_array, categorical=[0, 2, 4])
```

```python
print(kproto.cluster_centroids_)
```

```python
#1st center
kproto.cluster_centroids_[0][0],kproto.cluster_centroids_[1][0],
```

```python
#2nd center
kproto.cluster_centroids_[0][1],kproto.cluster_centroids_[1][1],
```

```python
#3rd center
kproto.cluster_centroids_[0][2],kproto.cluster_centroids_[1][2],
```

```python
predictions=[]
for c in clusters:
    predictions.append(c)
```

```python
predictions
```

```python
df['cluster']=predictions
```
