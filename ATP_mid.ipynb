# Importing the neccessary Libraries

```python
## Importing the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the Data Set

```python
df = pd.read_csv('C:/Users/sreemanta.kesh/Desktop/Clarity/ATP.csv')
```

## Narrowing the main Date Set in to three primary Data Frame(RF - Roger Federer, RN - Rafael Nadal, ND - Novak Djokovic)

```python
R_lose = df[df.loser_name == 'Roger Federer'] ##Rows where RF lost
R_win = df[(df.winner_name == 'Roger Federer')] ##Rows where RF won
R = [R_win, R_lose] 
RF = pd.concat(R)## Concating both win and loss to form the complete data Set for Federer

Rafa_lose = df[df.loser_name == 'Rafael Nadal']
Rafa_win = df[(df.winner_name == 'Rafael Nadal')] 
Rafa = [Rafa_lose, Rafa_win]
RN = pd.concat(Rafa)#Same as above

N_lose = df[df.loser_name == 'Novak Djokovic']
N_win = df[(df.winner_name == 'Novak Djokovic')] 
N = [N_lose, N_win]
ND = pd.concat(N)#Same as above
```

## Adding a Column 'Sets' to the data frames

```python
def element(str1):
    if '-' in str1:
        return str1.count('-')
    else:
        return 0    
RF['Sets'] = RF['score'].apply(element)
RN['Sets'] = RN['score'].apply(element)
ND['Sets'] = ND['score'].apply(element)
```

## Visualizing Individual Careers

```python


Total_Played = [RF.shape[0], RN.shape[0],  ND.shape[0]]##Total Matches
Total_Won = [R_win.shape[0], Rafa_win.shape[0],  N_win.shape[0]]## Total Matches WOn
y_pos1 = [0,4,8] ## Defining Position in the graph 
y_pos2 = [1,5,9]

plt.bar(y_pos1, Total_Played, width = 1, color = 'turquoise', edgecolor = 'black', capsize=7, label='Total Matches Played')
plt.bar(y_pos2, Total_Won, width = 1, color = 'green', edgecolor = 'black', capsize=7, label='Matches Won')

W1 = float(R_win.shape[0]/RF.shape[0])*100 
W1 = str(round(W1,2))  
W1 = '('+W1+')' 

W2 = float(Rafa_win.shape[0]/RN.shape[0])*100 
W2 = str(round(W2,2))
W2 = '('+W2+')'

W3 = float(N_win.shape[0]/ND.shape[0])*100  
W3 = str(round(W3,2))
W3 = '('+W3+')'

plt.xticks([0.5,4.5,8.5], ['Federer '+W1, 'Nadal '+W2, 'Djokovic '+W3])
plt.title('Career Statistics',size = 20, fontweight = 'bold')

plt.legend()
```
