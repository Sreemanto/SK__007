```python
import requests
import pandas as pd
```

```python
#Accessing the URL
r = requests.get('http://www.emlii.com/b6ec1de6/Events-Across-100-Years-That-Completely-Changed-The-World.html') 
```

```python
from bs4 import BeautifulSoup
```

```python
#Using the Beautiful Soup Object to understand the HTML
S = BeautifulSoup(r.text, 'html.parser')
```

```python
#checking the doc HTML
S
```

```python
#Each Event is wrapped under the tag <div class = 'article-inner-block'>

Events = S.find_all('div', attrs={'class':'article-inner-block'})
```

```python
#Total No. of Events
len(Events)
```

```python
# Each Event is wrapped in below pattern (Title, Image Link, Social Media lInks, Article, Source, Description)

# <p class="article-subtitle"> -- <div class="article-inner-block-image"> -- <div class="horizontal-share-block"> -- 
#            <div class="article-source"> -- <p class="article-inner-description">


#Extracting the 1st Event set

E1 = Events[0]
```

```python
# Displayin the 1st Event
E1
```

```python
# 1. Extracting the Title
E1.find('p').text[4:-7]
```

```python
# 2. Extracting the Year
E1.find('p').text[-5:-1]
```

```python
# 3. Extracting the Description
E1.contents[4].text
```

```python
# 4. Extracting the Image Link
'http://www.emlii.com' + str(E1.find('img')).split()[-1][5:-3]
```

```python
# Creating the set of events as a list of lists
List_of_Events = []
for Event in Events:
    Title = Event.find('p').text[4:-7]
    if Title == '':
        Year = 'NA'
    else:
        Year = Event.find('p').text[-5:-1]
    Description = Event.contents[4].text
    Image_Link = 'http://www.emlii.com' + str(Event.find('img')).split()[-1][5:-3]
    List_of_Events.append((Title, Year, Description, Image_Link))
```

```python
List_of_Events[0:4]
```

```python
#Now Creating a Dataset
df = pd.DataFrame(List_of_Events, columns=['Title','Year','Description','Image_Link'])
```
